```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

* Data Cleaning

```python
df.isnull().sum()
```

```python
# Dropping missing values

df.dropna(inplace=True)
```

```python
df[df.duplicated()]
```

```python
# deleting all deplicates

df.drop_duplicates(inplace=True)
```

```python
# Type Casting

df['id'] = df['id'].astype(object)
df['host_id'] = df['host_id'].astype(object)
```

```python
df.dtypes
```

# EDA
Data analysis

* Univariate Data Analysis

```python
# Checking for Outliers and removing them

df = df[df['price']<1500]

sns.boxplot(data=df, x="price")
```

```python
# Price Distribution

plt.figure(figsize=(8,5))
sns.histplot(data=df, x='price', bins=100)
plt.title("Price Distribution")
plt.ylabel('Frequency')
plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='availability_365')
plt.title("availability_365 Distribution")
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='bedrooms', bins=10)
plt.title("bedrooms Distribution")
plt.ylabel('Frequency')
plt.show()
```

```python
# ['price per bed'] feature Engineering

df['price per bed'] = df['price']/df['beds']
df.head()
```

```python
df.groupby(by='neighbourhood_group')['price'].mean()
```

```python
df.groupby(by='neighbourhood_group')['price per bed'].mean()
```

**Bi Variate Analysis** *One variable dependency on another vaiable*

```python
df.columns
```

```python
# Neighbourhood group and price relationship

sns.barplot(data=df, x='neighbourhood_group', y='price', hue='room_type')
```

```python
# Number of reviews and the price relationship

plt.figure(figsize=(8, 5))
plt.title("Neighbourhood Groups and Number of Reviews")
sns.scatterplot(data=df, x='number_of_reviews', y='price', hue='neighbourhood_group')
plt.show()
```

```python
sns.pairplot(data=df, vars=['price', 'minimum_nights', 'number_of_reviews', 'availability_365'], hue='room_type')
```

```python
# Geographical Distribution of Airbnb Listing

plt.figure(figsize=(10,7))
sns.scatterplot(data=df, x='longitude', y='latitude', hue='room_type')
plt.title("Geographical Distribution of Airbnb Listings")
plt.show()
```

```python
df.dtypes
```

```python
# Heatmap - Correlation of each numerical column

corr = df[['latitude', 'longitude', 'price',
       'minimum_nights', 'number_of_reviews',
       'reviews_per_month', 'calculated_host_listings_count',
       'availability_365', 'number_of_reviews_ltm','beds', 'price per bed']].corr()

corr

plt.figure(figsize=(10, 10))
sns.heatmap(data=corr, annot=True)
```
